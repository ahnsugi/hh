```python
# 데이터셋을 생성하기 위해 랜덤함수를 사용한다
# 배열 계산을 위해 numpy를 사용한다
import random
import numpy as np
```

```python
# 어린이, 학생, 어른 데이터 생성
# 키, 몸무게, 달리기속도로 구분한다
child = []
student = []
man = []

for i in range(50):
    child.append([random.randint(100, 140), random.randint(40, 60), random.randint(15, 25), 0])  # 어린이는 0번 그룹
    student.append([random.randint(130, 170), random.randint(50, 80),random.randint(20, 35), 1]) # 학생은 1번 그룹
    man.append([random.randint(150, 190), random.randint(60, 100),random.randint(20, 30), 2])    # 어른은 2번 그룹
print(child)
print(student)
print(man)
```

```python
#거리계산함수 정의
def distance(x, y):
    return (np.sqrt(pow((x[0]-y[0]), 2) + pow((x[1]-y[1]),2)) + pow((x[2]-y[2]),2))
```

```python
#임의의 데이터를 입력받는다
new = []
new.append( int(input("키를 입력하세요: ")) )
new.append( int(input("몸무게를 입력하세요: ")) )
new.append(int(input("달리기속도를 입력하세요: ")) )
new.append(-1)  # 구분되지 않은 경우 그룹을 -1로 설정한다
print(new)
```

```python
# 입력데이터와 데이터셋 간의 거리를 측정하여 result 리스트에 넣는다
result = []
for i in range(len(child)):
    result.append([distance(new, child[i]), child[i][3]])
for i in range(len(student)):
    result.append([distance(new, student[i]), student[i][3]])
for i in range(len(man)):
    result.append([distance(new, man[i]), man[i][3]])
    
# 결과를 가까운 순서대로 정렬한다
result.sort()
print (result)
```

```python
# target에 따라 어린이/학생/어른을 판단하는 함수
def judge(result, target):
    cnt= [0,0,0] #카운트배열, 앞부터 어린이, 학생, 어른 수를 셈
    for i in range(target):
        if(result[target][1] == 0):    # 리스트 그룹이 0이면
            cnt[0]+=1  #어린이 숫자를 증가시킨다
        elif(result[target][1] == 1):   # 그룹이 1번이면
            cnt[1]+=1  # 학생 숫자를 증가시킨다
        else: #어른 그룹이면
            cnt[2]+=1
    # 아이, 어른, 학생 결과를 출력한다
    print("k =", target, "일때 입력한 값은")
    if(max(cnt)== cnt[0]): #어린이 값이 최대이면
        print("어린이입니다")
        return 0
    elif(max(cnt) == cnt[1]):  # 학생 수가 최대이면
        print("학생입니다")
        return 1
    else: #어른이 최대이면
        print("어른입니다")
        return 2
    
        

```
