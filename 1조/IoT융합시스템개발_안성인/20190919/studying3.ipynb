```python
import tensorflow as tf
print(tf.__version__)
import tensorflow.keras.utils as utils
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
import numpy as np
dataset = np.loadtxt("C:/Users\IoT08/Documents/20190919/pima-indians-diabetes.csv", delimiter=",")
# print(dataset)
# print(dataset[0,0:8])
# print(dataset[0,8])
```

```python
x_train = dataset[:600, 0:8]
y_train = dataset[:600, 8]
x_val = dataset[600:700, 0:8]
y_val = dataset[600:700, 8]
x_test = dataset[:700, 0:8]
y_test = dataset[:700, 8]
```

```python
model = Sequential()
model.add(Dense(units=20, input_dim=8, activation='relu'))
model.add(Dense(units=10, activation = 'relu'))
model.add(Dense(units=1, activation = 'sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(patience=10000)

#딥러닝 실행
hist = model.fit(x_train, y_train, epochs=100, batch_size=10, verbose=1, validation_data=(x_val, y_val), callbacks=[early_stopping])

#결과 출력
#모델 학습 과정 표시하기
%matplotlib inline
import matplotlib.pyplot as plt

fig, loss_ax = plt.subplots()

acc_ax = loss_ax.twinx()

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'r', label='val loss')

acc_ax.plot(hist.history['accuracy'], 'b', label='train acc')
acc_ax.plot(hist.history['val_accuracy'], 'g', label='val acc')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuracy')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
```
