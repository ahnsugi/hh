```python
print(X_train.shape)
print(Y_train.shape)
```

```python
# 라벨링 전환
Y_train = utils.to_categorical(Y_train,19)
Y_val = utils.to_categorical(Y_val,19)
```

```python
print(Y_train[1])
```

```python
model = Sequential()
```

```python
model.add(Dense(units=38, input_dim=1, activation='elu'))
model.add(Dense(units=19,  activation='softmax'))
```

```python
# 3. 모델 엮기
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```python
# 4. 모델 학습시키기
hist = model.fit(X_train, Y_train,
                 epochs=500,
                 batch_size=10,
                 verbose=1,
                 validation_data=(X_val, Y_val))
```

```python
# 5. 모델 학습 과정 표시하기
%matplotlib inline
import matplotlib.pyplot as plt

fig, loss_ax = plt.subplots()

acc_ax = loss_ax.twinx()

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'r', label='val loss')

acc_ax.plot(hist.history['accuracy'], 'b', label='train acc')
acc_ax.plot(hist.history['val_accuracy'], 'g', label='val acc')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuracy')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
```

```python
width = 28
height = 28
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(60000, width*height).astype('float32') / 255.0
x_test = x_test.reshape(10000, width*height).astype('float32') / 255.0
```

```python
x_val = x_train[50000:]
y_val = y_train[50000:]
x_train = x_train[:50000]
y_train = y_train[:50000]
```

```python
y_train = utils.to_categorical(y_train)
y_val = utils.to_categorical(y_val)
y_test = utils.to_categorical(y_test)
```

```python
# 3. 모델 엮기
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
# 카테고리 방식, 못할때 매로 맞을것인가? 잘할때 사탕을 줄 것인가, 성능지표 정확도 판단하겠다.

# 4. 모델 학습시키기
hist = model.fit(X_train, Y_train, epochs=10, batch_size=30, verbose=1, validation_data=(X_val, Y_val))
# 입력, 출력, 반복횟수, 한번에 몇 문제를 풀것이냐, 얼만큼 설명해줄것인가
```

```python
# 5. 모델 학습 과정 표시하기
%matplotlib inline
import matplotlib.pyplot as plt
# 주피터 노트북 안에서 그래프 표현

fig, loss_ax = plt.subplots()

acc_ax = loss_ax.twinx()
# 두개의 그래프를 섞어 표현

loss_ax.plot(hist.history['loss'], 'y', label='train loss')            # 노란색 훈련 데이터 틀린 갯수
loss_ax.plot(hist.history['val_loss'], 'r', label='val loss')          # 빨간색 검증 데이터 틀린 갯수

acc_ax.plot(hist.history['accuracy'], 'b', label='train acc')          # 파란색 훈련 데이터 정확도(맞힌 갯수)
acc_ax.plot(hist.history['val_accuracy'], 'g', label='val acc')        # 초록색 검증 데이터 정확도(맞힌 갯수)
# 최종 loss와 accuracy를 그래프로 할 것이다. loss와 accuracy는 반비례

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuray')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
```
