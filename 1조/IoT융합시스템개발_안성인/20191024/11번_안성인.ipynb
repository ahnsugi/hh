# 1. 파이썬 활용 능력
## - 구구단을 2단부터 9단까지 출력하세요.
## (단, 정수 하나를 입력 받고 입력 받은 정수는 제외하고 출력하세요.
##  예 : 숫자 3 입력 -> 2, 4, 5, 6, 7, 8, 9단 출력)

```python
num = int(input("제외하고 싶은 단을 적어주세요."))
print("{}단은 제외합니다.\n".format(num))
for j in range(2, 10):#2단부터 9단까지 출력
    if j == num:
        continue # 제외할 단일 때 아래 실행 안 하도록 하는 문장
    print("====={}단=====".format(j))
    for i in range(1, 10): # 해당 단의 내용 출력
        print("{} X {} = {:2d}".format(j,i,j*i))
    print("=============\n")
```

# 2. 기계학습 이해
## - KNN 알고리즘을 구현하세요.
##  (k가 3, 5, 7일 때 어린이, 청소년, 어른으로 구분하는 모델, 특징은 최소 3가지 이상)

```python
import numpy as np # numpy : 다차원 함수 
import random # 랜덤함수
import matplotlib.pyplot as plt # 시각화로 표현함. 애칭을 pit로 지정
get_ipython().run_line_magic('matplotlib', 'inline')
c = [] # 어린이
t = [] # 청소년
a = [] # 어른

for i in range(50): # 첫번째 값은 나이, 두번째 값은 발사이즈, 세번째 값은 피부탄력
    c.append([random.randint(1, 13)*10, random.randint(76, 220), random.randint(1, 2)*10, 0])
    t.append([random.randint(14, 19)*10, random.randint(210, 280), random.randint(2, 5)*10, 1])
    a.append([random.randint(20, 100)*10, random.randint(240, 300), random.randint(5, 40)*10, 2])
    # 나이와 피부 탄력에 10씩 곱한 이유는 값의 크기 차이 때문에 우선순위가 발생하지 않도록 값을 올렸습니다.
def distance(x,y):# 두점 사이의 거리
    return np.sqrt(pow((x[0]-y[0]), 2) + pow((x[1] - y[1]), 2) + pow((x[2] - y[2]), 2))

def knn(x, y, k): # knn 함수
    result = []
    cnt = 0
    for i in range(len(y)):
        result.append([distance(x,y[i]),y[i][3]])
    result.sort()
    for i in range(k-1):
        if(result[i][1] == 2):
            cnt += 4 # 값이 어른의 값과 유사할 때 해당 값만큼 증가
        elif(result[i][1] == 1):
            cnt += 2 # 값이 청소년의 값과 유사할 때 해당 값만큼 증가
    if(cnt > k*2):
        print("당신은 어른입니다.")
    elif(cnt <= k*2 and cnt > k):
        print("당신은 청소년입니다.")
    else:
        print("당신은 어린이입니다.")

for i in range(50): 
    plt.plot(c[i][0], c[i][1], c[i][2], marker='o', color = 'red')
    plt.plot(t[i][0], t[i][1], t[i][2], marker='o', color = 'blue')
    plt.plot(a[i][0], a[i][1], a[i][2], marker='o', color = 'green')

new = [] #새로운 좌표

num = random.randint(1,3)
if(num == 1):   # 어린이일때
    new.append([random.randint(1, 13)*10, random.randint(76, 220), random.randint(1, 2)*10])
elif(num == 2): # 청소년일떄
    new.append([random.randint(14, 19)*10, random.randint(210, 280), random.randint(2, 5)*10])
else:           # 어른일때
    new.append([random.randint(20, 100)*10, random.randint(240, 300), random.randint(5, 40)*10])

age = input("나이를 입력해주세요. ")
footSize = input("발사이즈를 입력해주세요. ")
skinElasticity = input("피부탄력을 입력해주세요. ")
new = [int(age)*10,int(footSize), int(skinElasticity)*10]
for i in 3, 5, 7:
    knn(new,c+t+a,i)
```

# 3. 딥러닝의 이해
## - CNN을 활용해서 MNIST 손글씨 데이터를 학습 및 평가해주세요.
##  (정확도가 93% 이상이어야 함)

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import sys
import tensorflow as tf
import tensorflow.keras as keras
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
np.random.seed(7)

print('Python version : ', sys.version)
print('TensorFlow version : ', tf.__version__)
print('Keras version : ', keras.__version__)
print()

img_rows = 28
img_cols = 28

# 1. 데이터 생성하기
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

input_shape = (img_rows, img_cols, 1)
x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)

x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
print()

batch_size = 128
num_classes = 10
epochs = 12

y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

# 2. 모델 구성하기
model = Sequential()
model.add(Conv2D(32, kernel_size=(5, 5), strides=(1, 1), padding='same',
                 activation='relu',
                 input_shape=input_shape))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model.add(Conv2D(64, (2, 2), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(1000, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))
model.summary()

# 3. 모델 학습과정 설정하기
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 4. 모델 학습시키기
hist = model.fit(x_train, y_train,
                 batch_size=batch_size,
                 epochs=epochs,
                 verbose=1, 
                 validation_data=(x_test, y_test))
print()
# 5. 모델 평가하기
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
