# while

```python
front = 1
back = 2
while front<10:
    print("{} Dan".format(front))
    print("=========================================")    
    front= front+1  
    while back<10:
        print("{} x{} = {}".format(front, back, front*back))
        back = back+1
        
    back = 1
    print("\n")
```

```python
l = []
count = 0
for i in range(0,100):   
    a=int(random.random()*100)
    if(0<a and a<46):
        if(a not in l):
            l.append(a)
            count +=1
        if(count==6):
            break
print(l)



```

```python
# lotto 프로그램

lotto = [12,17,28,41,43,44]
limit = 0
num = [0,0,0,0,0]

while True:
    rand = random.sample(range(1,45),6)
    count =0
    bonus =0
    for i in rand:
        if (i in lotto):
            count +=1
        if(rand== 25):
            bonus =1
    limit +=1
    if(limit % 10000 ==0):
        print(limit,"번째 복권 구매 중")
    if(count == 6):
        print(limit, "번째 복권 1등 당첨!![",rand,"]")
        num[4]+=1
        break
    elif(count == 5):
        if(bonus in rand):
            print(limit, "번째 복권 2등 당첨!![",rand,"]")
            num[3]+=1
        else:
            print(limit, "번째 복권 3등 당첨!![",rand,"]")
            num[2]+=1
    elif(count == 4):
        print(limit, "번째 복권 4등 당첨!![",rand,"]") 
        num[1]+=1
    elif(count == 3):
        print(limit, "번째 복권 5등 당첨!![",rand,"]")
        num[0]+=1
    else:
        continue

print("복권 구입 개수:",num)
print("1등:",num[4])
print("2등:",num[3])
print("3등:",num[2])
print("4등:",num[1])
print("5등:",num[0])

        
```
