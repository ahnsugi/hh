```python
#기계학습 이해
import numpy as np  #ndarray 생성
import random  #random 함수 이용 
import matplotlib.pyplot as plt  #그래프를 그리기 위한 헤더
%matplotlib inline
c =[] #어린이 3       남성기준
t = [] #청소년 5
a = [] #어른 7
for i in range(50):  #50개의 랜덤 데이터 생성
    c.append([random.randint(100,160),random.randint(20,60),random.randint(200,210),3])
    t.append([random.randint(150,180),random.randint(40,100),random.randint(177,188),5])
    a.append([random.randint(160,200),random.randint(50,150),random.randint(100,167),7])
                #키                   # 몸무게                   #들을수 있는 주파수 100hz = 1hz로 따짐
        
#점 사이의 거리를 구하는 함수
def distance(x,y):
    return np.sqrt(pow((x[0]-y[0]), 2) + pow((x[1] - y[1]), 2)+ pow((x[2] - y[2]), 2))
result = []
CS=0
TS=0
AS=0
k=0
new = [0,0,0,0,0,0] #데이터 값을 받음
aa=[50,50,50,50] #데이터의 개수


new[0] = int(input("키를 입력해주세요:"))
new[1] = int(input("몸무게를 입력해주세요:"))
new[2] = int(input("가청 주파수를 입력해주세요:"))
k = int(input("k의 값을 입력해주세요:")) # k만큼 비교

for i in range(aa[0]):  #그래프 그리기
    plt.plot(c[i][0],c[i][1],c[i][2], marker='o', color='red')
    plt.plot(t[i][0],t[i][1],t[i][2], marker='o', color='blue')
    plt.plot(a[i][0],a[i][1],a[i][2], marker='o', color='green')
        
        
plt.plot(new[0],new[1],new[2], marker='x', color='black')  #입력 데이터 그래프 표시

for i in range(aa[1]):
    result.append([distance(new,c[i]),c[i][3]])
for i in range(aa[2]):
    result.append([distance(new,t[i]),t[i][3]])
for i in range(aa[3]):
    result.append([distance(new,a[i]),a[i][3]])

result.sort()  #거리에 따라 정렬
   
for i in range(k):  
    if result[i][1] == 3:   #어린이면 카운트
        CS+=1
    elif result[i][1] == 5: #청소년이면 카운트
        TS+=1
    else:                   #어른이면 카운트
        AS+=1
if(CS > (k/2)):  #카운트 개수가 더 많으면 
    print("당신은 어린이입니다")
    c.append([new[0],new[1],new[2],3])
    aa[1]+=1
elif(TS > (k/2)):
    print("당신은 청소년입니다")
    t.append([new[0],new[1],new[2],5])
    aa[2]+=1          
else:  #카운트 개수가 더 작으면
    print("당신은 어른입니다")
    a.append([new[0],new[1],new[2],7])
    aa[3]+=1
```

```python
import tensorflow.keras.utils as utils
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
import numpy as np

width = 28
height = 28
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train=x_train.reshape(60000, width*height).astype('float32')/255.0  # 28*28 정사각형 데이터를 1줄짜리 데이터로 바꾼다
x_test = x_test.reshape(10000, width*height).astype('float32')/255.0  # 정규화하기 위해 255로 나눈다

x_val = x_train[50000:]
y_val = y_train[50000:]
x_train = x_train[:50000]
y_train = y_train[:50000]
y_train = utils.to_categorical(y_train)
y_val = utils.to_categorical(y_val)
y_test = utils.to_categorical(y_test)

model = Sequential()
model.add(Dense(units=128, input_dim=width*height, activation='elu'))
model.add(Dense(units=10, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
              #loss와optimizer 명령어가 중요
    
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(patience=10)#조기종료 콜백함수 정의
hist = model.fit(x_train, y_train, epochs=50, batch_size=30, verbose=1, validation_data=(x_val, y_val))


loss_and_metrics = model.evaluate(x_test, y_test, batch_size=1)

print('')
print('loss : ' + str(loss_and_metrics[0]))
print('accuray : ' + str(loss_and_metrics[1]))
```
