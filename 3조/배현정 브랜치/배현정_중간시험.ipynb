# 4. 특징이 최소 5개 이상인 동물 분류기를 만들어주세요. (10)
### - 특징 5개 (예, 평균 크기, 육식 여부 등)
### - 분류할 동물은 최소 4마리 이상 (호랑이, 개, 원숭이 등)
### - knn 혹은 kmeans 알고리즘을 사용하세요.

```python
import numpy as np # ndarray 생성
import random # 랜덤 함수 사용을 위한 절차
tiger = []
dog = []
monkey = []
rabbit = []
cow = []

# 평균 크기(cm)
# 평균 몸무게(kg)
# 육식여부 (0, 1, 2)
# 가축으로 키우는지 여부 (0, 1)
# 사람이 먹을 수 있는지 여부 (0, 1)
# 동물 구분값 호랑이 - 0, 개 - 1, 원숭이 - 2, 토끼 - 3, 소 - 4
for i in range(50): # 하위 내용을 i값이 1씩 증가하면서 50번 반복
    tiger.append([random.randint(195, 300), random.randint(140,230), 1, 0, 0, 0] )
    # 195 - 300cm,140 - 230kg
    dog.append([random.randint(55,94), random.randint(22,28), 2, 1, 1, 1])
    # 55 - 94cm, 22 - 28kg
    monkey.append([random.randint(35,40), random.randint(2,3),0, 0, 0, 2])
    # 35 - 40cm, 2 - 3kg
    rabbit.append([random.randint(45,54), random.randint(1,2), 0, 1, 1, 3])
    # 45 - 54cm, 1 - 2kg
    cow.append([random.randint(210,290), random.randint(300,420), 0, 1, 1, 4])
    # 210 - 290cm, 300 - 420kg

def distance(x,y):
    # 4차원에서 두 점 사이의 거리를 구하는 함수
    return np.sqrt(pow((x[0]-y[0]),2)+pow((x[1]-y[1]),2)+pow((x[2]-y[2]),2)+pow((x[3]-y[3]),2)+pow((x[4]-y[4]),2))

def knn(x,y,k):
    result=[]
    cnt=[0,0,0,0,0]
    # cnt[0]=0 호랑이를 카운트 하기 위한 변수
    # cnt[1]=0 개를 카운트 하기 위한 변수
    # cnt[2]=0 원숭이를 카운트 하기 위한 변수
    # cnt[3]=0 토끼를 카운트 하기 위한 변수
    # cnt[4]=0 소를 카운트 하기 위한 변수
    for i in range(len(y)): # 하위 내용을 i값이 1값이 1씩 증가하면서 y 의 갯수 만큼 반복
        result.append([distance(x,y[i]),y[i][5]]) 
        # "x와 y의 두 점 사이의 거리를 구한 값"과 "동물 구분 값"을 리스트에 추가
    result.sort() # result 를 정렬
    for i in range(k): # 하위 내용을 i값이 1값이 1씩 증가하면서 k 의 갯수 만큼 반복
        if (result[i][1] == 0):
            cnt[0] +=1
        if (result[i][1] == 1):
            cnt[1] +=1
        if (result[i][1] == 2):
            cnt[2] +=1
        if (result[i][1] == 3):
            cnt[3] +=1
        if (result[i][1] == 4):
            cnt[4] +=1
    if (max(cnt)==cnt[0]): print("새로 입력한 값에 대한 동물은 호랑이입니다.")
    elif (max(cnt)==cnt[1]): print("새로 입력한 값에 대한 동물은 개입니다.")
    elif (max(cnt)==cnt[2]): print("새로 입력한 값에 대한 동물은 원숭이입니다.")
    elif (max(cnt)==cnt[3]): print("새로 입력한 값에 대한 동물은 토끼입니다.")
    else: print("새로 입력한 값에 대한 동물은 소입니다.")
        
cm = input("동물의 몸길이를 입력해주세요 : ")
kg = input("동물의 몸무게를 입력해주세요 : ")
eating = input("육식 여부를 입력해주세요 (초식 0, 육식 1, 잡식 2) : ")
home = input("가축으로 키우는지 여부를 입력해주세요 (키우지 않는다 0, 키운다 1) : ")
eaten = input("사람이 먹을 수 있는지 여부를 입력해주세요 (먹지 않는다 0, 먹는다 1) : ")
num = input("K를 입력해주세요 : ")
new = int(cm), int(kg), int(eating), int(home), int(eaten)
    
knn(new,tiger+dog+monkey+rabbit+cow,int(num))
```
