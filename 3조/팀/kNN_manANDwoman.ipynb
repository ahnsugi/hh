## KNN 알고리즘을 사용해서 남자/여자 분류하기 

### 조건 

1. 특성은 최소 네 개 이상 -> 키, 몸무게 등등등

2. 최종 결과물은 k가 3, 5, 7, 9, 11 개 일 때 각각 랜덤한 사람 3명의 결과 

3. 각 조마다 각자 branch에서 작업하고, 최종 결과물은 팀원 중 한명이 master의 각 조 파일에 업로드 

```python
import numpy as np # ndarray 생성

import random # 랜덤 함수 사용을 위한 절차
r = [] # 여자 1
b = [] # 남자 0

for i in range(50): # 하위 내용을 i값이 1씩 증가하면서 50번 반복
    r.append([random.randint(40,70), random.randint(140, 180), random.randint(200, 260), random.randint(50, 60), 1])
    # [ 몸무게 40 ~ 70 까지 , 키 140 ~ 180 까지 , 발 사이즈 200 ~ 260 까지, 머리 둘레 50 ~ 60 까지, 1 - 여자 ] r에 추가
    b.append([random.randint(60,90), random.randint(160, 200), random.randint(240, 300), random.randint(55, 65), 0])
    # [ 몸무게 60 ~ 90 까지 , 키 160 ~ 200 까지 , 발 사이즈 240 ~ 300 까지, 머리 둘레 55 ~ 65 까지, 0 - 남자 ] b에 추가

def distance(x,y):
    # 4차원에서 두 점 사이의 거리를 구하는 함수
    return np.sqrt(pow((x[0]-y[0]),2)+pow((x[1]-y[1]),2)+pow((x[2]-y[2]),2)+pow((x[3]-y[3]),2))

def knn(a,b,c,p,k): # 첫 번째 사람 -> a / 두 번째 사람 -> b / 세 번째 사람 -> c / p -> r + b / k -> 남녀구분
    result1=[] # 첫 번째 사람 분류를 위한 리스트
    result2=[] # 두 번째 사람 분류를 위한 리스트
    result3=[] # 세 번째 사람 분류를 위한 리스트
    cnt1=0 # 첫 번째 사람 분류를 위한 카운트 변수
    cnt2=0 # 두 번째 사람 분류를 위한 카운트 변수
    cnt3=0 # 세 번째 사람 분류를 위한 카운트 변수
    for i in range(len(p)): # 하위 내용을 i값이 1값이 1씩 증가하면서 p = r + b 의 갯수 만큼 반복
        result1.append([distance(a,p[i]),p[i][4]]) 
        # "첫 번째 사람 정보와 p의 두 점 사이의 거리를 구한 값"과 "남,녀 구분 값"을 리스트에 추가
        result2.append([distance(b,p[i]),p[i][4]]) 
        # "두 번째 사람 정보와 p의 두 점 사이의 거리를 구한 값"과 "남,녀 구분 값"을 리스트에 추가
        result3.append([distance(c,p[i]),p[i][4]])
        # "세 번째 사람 정보와 p의 두 점 사이의 거리를 구한 값"과 "남,녀 구분 값"을 리스트에 추가
    result1.sort() # result1 을 정렬
    result2.sort() # result2 를 정렬
    result3.sort() # result3 를 정렬
    for i in range(k): # 하위 내용을 i값이 1값이 1씩 증가하면서 k 의 갯수 만큼 반복
        if (result1[i][1] == 1): # rsult1의 남녀구분을 한 결과(k)가 1 이면 하위 내용을 실행
            cnt1 +=1 # cnt1 을 1 증가
        if (result2[i][1] == 1): # rsult2의 남녀구분을 한 결과(k)가 1 이면 하위 내용을 실행
            cnt2 +=1 # cnt2 를 1 증가
        if (result3[i][1] == 1): # rsult3의 남녀구분을 한 결과(k)가 1 이면 하위 내용을 실행
            cnt3 +=1 # cnt3 을 1 증가
    if (cnt1 > (k/2)): # 위에서 증가시킨 cnt1이 과반수(k/2) 보다 크면 하위 내용을 실행
        print("첫 번째 입력한 사람은 여자입니다.") # 출력
    else: # 아니면
        print("첫 번째 입력한 사람은 남자입니다.") # 출력
    if (cnt2 > (k/2)): # 위에서 증가시킨 cnt2가 과반수(k/2) 보다 크면 하위 내용을 실행
        print("두 번째 입력한 사람은 여자입니다.") # 출력
    else: # 아니면
        print("두 번째 입력한 사람은 남자입니다.") # 출력
    if (cnt3 > (k/2)): # 위에서 증가시킨 cnt3이 과반수(k/2) 보다 크면 하위 내용을 실행
        print("세 번째 입력한 사람은 여자입니다.") # 출력
    else: # 아니면
        print("세 번째 입력한 사람은 남자입니다.") # 출력
        
for i in range(1,6): # k가 3, 5, 7, 9 ,11 (5번) 전부 비교하려고 range를 1 ~ 6 까지 했습니다.
    weight = input("첫 번째 사람의 몸무게를 입력해주세요 : ")
    height = input("첫 번째 사람의 키를 입력해주세요 : ")
    shoe_size = input("첫 번째 사람의 신발 사이즈를 입력해주세요 : ")
    hat_size = input("첫 번째 사람의 모자 사이즈를 입력해주세요 : ")
    num = input("K를 입력해주세요 : ")
    new1 = int(weight), int(height), int(shoe_size), int(hat_size)
    # new1에 내가 입력한 weight(몸무게), height(키), shoe_size(발 사이즈), hat_size(머리 둘레) 가 입력될 것입니다.

    weight = input("두 번째 사람의 몸무게를 입력해주세요 : ")
    height = input("두 번째 사람의 키를 입력해주세요 : ")
    shoe_size = input("두 번째 사람의 신발 사이즈를 입력해주세요 : ")
    hat_size = input("두 번째 사람의 모자 사이즈를 입력해주세요 : ")
    num = input("K를 입력해주세요 : ")
    new2 = int(weight), int(height), int(shoe_size), int(hat_size)
    # new2에 내가 입력한 weight(몸무게), height(키), shoe_size(발 사이즈), hat_size(머리 둘레) 가 입력될 것입니다.

    weight = input("세 번째 사람의 몸무게를 입력해주세요 : ")
    height = input("세 번째 사람의 키를 입력해주세요 : ")
    shoe_size = input("세 번째 사람의 신발 사이즈를 입력해주세요 : ")
    hat_size = input("세 번째 사람의 모자 사이즈를 입력해주세요 : ")
    num = input("K를 입력해주세요 : ")
    new3 = int(weight), int(height), int(shoe_size), int(hat_size)
    # new3에 내가 입력한 weight(몸무게), height(키), shoe_size(발 사이즈), hat_size(머리 둘레) 가 입력될 것입니다.
    
    knn(new1,new2,new3,r+b,int(num)) # 위에 있는 knn 함수 참조
```
