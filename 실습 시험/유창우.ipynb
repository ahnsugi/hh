# 로또

```python
l = [] # list 생성
count = 0 # 변수 생성
while count < 6: # 로또 6자리
    tmp = int(random.random()*100) # 10의 자리 랜덤 값 tmp에 저장
    if((tmp != 0) & (tmp < 46)): # 저장된 tmp 값들 중 로또(1부터 45까지 숫자)에 해당하는 것들이 나오면
        if(tmp not in l): # tmp가 l 안에 없으면 - 로또 중복 값 제거 목적
            l.append(tmp) # l에 랜덤 값 tmp 추가
            count += 1 # count 1 추가
            
print(l) # l 출력
```

# 로또 6자리 전부 맞추려면 복권을 얼마나 구매해야하는가?

```python
lotto = [12,17,28,41,43,44] # lotto 번호를 12,17,28,41,43,44라 가정.
limit = 0 # 변수
while True: # 무한 반복
    tmp = random. sample(range(1,45),6) # tmp에 6번 1 ~ 45 랜덤 값 저장
    count = 0 # 변수
    for i in tmp: # tmp 만큼 반복이니 6번 반복할 것임
        if (i in lotto): # 만약 위에서 내려온 i 즉 tmp가 lotto 안에 값과 같은 값이면
            count += 1 # count 값을 1 증가
    limit += 1 # 이 구간에 올때마다 limit 값 1 증가
    if(limit % 10000 == 0): # 그렇게 증가한 limit를 1만 단위에 도달할 때마다
        print(limit, " 번째 복권 구매 중") # 출력
    if(count == 6): # 위 if 문에서 증가한 count값이 6이 되면 아래 실행
        print(limit, "번째 당첨 !! [ ", tmp, "]") # 출력
        break
```

# 특징을 최소 5개 이상인 동물 분류기를 만들어주세요.
## 특징 5개 (예: 크기, 수명, 먹이, 무게, 속도)
## 분류할 동물은 최소 4마리 이상 (고등어, 구렁이, 독수리, 호랑이 등)
## knn 혹은 kmeans 알고리즘을 사용하세요.

```python
import numpy as np # ndarray 생성

import random # 랜덤 함수 사용을 위한 절차
Fish = [] # 고등어 0
Snake = [] # 구렁이 1
Bird = [] # 독수리 2
Tiger = [] # 호랑이 3

for i in range(50): # 하위 내용을 i값이 1씩 증가하면서 50번 반복
    Fish.append([random.randint(3,4), random.randint(1,3), 0, random.randint(1,3), random.randint(6,7), 0])
    # [ 크기 3~4, 수명 1~3, 먹이 0, 몸무게 1~3, 속도 6~7] Fish 에 추가
    Snake.append([random.randint(15,18), random.randint(1,3), random.randint(2,4), random.randint(7,14), random.randint(5,6), 1])
    # [ 크기 15~18, 수명 1~3, 먹이 2~4, 몸무게 7~14, 속도 5~6] Snake 에 추가
    Bird.append([random.randint(9,12), random.randint(2,3), random.randint(0,3), random.randint(6,14), random.randint(12,20), 2])
    # [ 크기 9~12, 수명 2~3, 먹이 0~3, 몸무게 6~14, 속도 12~20] Bird 에 추가
    Tiger.append([random.randint(18,40), random.randint(1,2), random.randint(2,3), random.randint(7,30), random.randint(4,7), 3])
    # [ 크기 18~40, 수명 1~2, 먹이 2~3, 몸무게 7~30, 속도 4~7] Tiger 에 추가


def distance(x,y):
    # 5차원에서 두 점 사이의 거리를 구하는 함수
    return np.sqrt(pow((x[0]-y[0]),2)+pow((x[1]-y[1]),2)+pow((x[2]-y[2]),2)+pow((x[3]-y[3]),2)+pow((x[4]-y[4]),2))

def knn(x,y,k): # x -> new 동물 특징 5가지 / y -> 새 동물의 크기+수명+먹이종류+몸무게+속도 / k -> 동물구분
    result=[] # 동물 분류를 위한 리스트
    cnt=[0,0,0,0] # 변수 선언
    for i in range(len(y)): # 하위 내용을 i값이 1씩 증가하면서 y의 갯수 만큼 반복
        result.append([distance(x,y[i]),y[i][5]]) 
        # "x와 y의 두 점 사이의 거리를 구한 값"과 "동물 구분 값"을 리스트에 추가
    result.sort() # result 를 정렬
    for i in range(k): # 하위 내용을 i값이 1값이 1씩 증가하면서 k 의 갯수 만큼 반복
        if (result[i][1] == 0): # rsult[i][1]이 0일 때마다 cnt0을 1 증가시켜 고등어
            cnt[0] +=1
        if (result[i][1] == 1): # rsult[i][1]이 1일 때마다 cnt1을 1 증가시켜 구렁이
            cnt[1] +=1
        if (result[i][1] == 2): # rsult[i][1]이 2일 때마다 cnt2을 1 증가시켜 독수리
            cnt[2] +=1
        if (result[i][1] == 3): # rsult[i][1]이 3일 때마다 cnt3을 1 증가시켜 호랑이
            cnt[3] +=1
    if (max(cnt)==cnt[0]): print("고등어입니다.")
    elif (max(cnt)==cnt[1]): print("구렁이입니다.")
    elif (max(cnt)==cnt[2]): print("독수리입니다.")
    else: print("호랑이입니다.")
        
Size = input("크기 3 ~ 40 입력 : ")
Age = input("수명 1 ~ 3 입력: ")
Eat = input("먹이 종류 0 ~ 4 입력 : ")
Weight = input("무게 1 ~ 30 입력 : ")
Speed = input("속도 4 ~ 20 입력 : ")
num = input("홀수 입력: ")
new = int(Size), int(Age), int(Eat), int(Weight), int(Speed)
    
knn(new,Fish+Snake+Bird+Tiger,int(num)) # 위에 있는 knn 함수 참조
```
